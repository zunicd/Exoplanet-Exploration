Let's define a small helper funtcion which displays count and percentage per class of the target feature.

```python
# Function to display count and percentage per class of target feature
def class_count(a):
    counter=Counter(a)
    kv=[list(counter.keys()),list(counter.values())]
    dff = pd.DataFrame(np.array(kv).T, columns=['koi_disposition','Count'])
    dff['Count'] = dff['Count'].astype('int64')
    dff['%'] = round(dff['Count'] / a.shape[0] * 100, 2)
    return dff.sort_values('Count',ascending=False)
```

```python
# Let's use the function
class_count(y)
```

# Pre-processing

### Scaling

Scale the data using the MinMaxScaler

```python
# Scale your data
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



### Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=10, max_iter=1000)
lrc.fit(X_train_scaled, y_train)
```

```python
LRC_Train_Score = lrc.score(X_train_scaled, y_train)
LRC_Test_Score = lrc.score(X_test_scaled, y_test)                        
print(f"Training Data Score: {LRC_Train_Score}")
print(f"Testing Data Score: {LRC_Test_Score}")
```

```python
# Predictions
pred = lrc.predict(X_test_scaled)
pred
```

```python
pd.DataFrame({"Prediction": pred, "Actual": y_test}).reset_index(drop=True)
```

### Validation

Let' start with confusion matrix

```python
cm = confusion_matrix(y_test, pred)
print(cm)
```

Our data is sltghtly imbalanced so we will use normalized confusion matrix as well.

```python
# Normalized confusion matrix
cm = np.around(cm / cm.sum(axis=1)[:, np.newaxis], 2)
print(cm)
```

It is not so easy to get the storu from just numbers. 

This is why hy we will create a heatmap plot for both matrices sid e by side.

```python
# Function to display confusion matrix and its normalized copy side by side
def plot_conf_mat_w_and_wo_norm(model):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))
    f.suptitle(model, fontsize=14)
    f.subplots_adjust(top=0.85, wspace=0.2)

    # confusion matrix without normalization
    mat = confusion_matrix(y_test, pred)
    sns.heatmap(mat,
                annot=True,
                annot_kws=dict(fontsize=14),
                fmt='d',
                cbar=True,
                square=True,
                cmap='Reds',
                ax=ax1)

    ax1.set_xticklabels(labels=target_names, size=9)
    ax1.set_yticklabels(labels=target_names, size=9, va='center')
    ax1.set_title('Confusion Matrix w/o Normalization')
    ax1.set_xlabel('Predicted Labels', size=12)
    ax1.set_ylabel('True Labels', size=12)

    # normalized confusion matrix
    matn = mat / mat.sum(axis=1)[:, np.newaxis]
    sns.heatmap(matn,
                annot=True,
                annot_kws=dict(fontsize=14),
                fmt='.2f',
                cbar=True,
                square=True,
                cmap='Reds',
                vmin = 0,
                vmax = 1,
                ax=ax2)

    ax2.set_xticklabels(labels=target_names, size=9)
    ax2.set_yticklabels(labels=target_names, size=9, va='center')
    ax2.set_title('Normalized Confusion Matrix')
    ax2.set_xlabel('Predicted Labels', size=12)
    ax2.set_ylabel('True Labels', size=12)

    plt.show()
```

```python
model = 'Logistic Regression'
plot_conf_mat_w_and_wo_norm(model)
```

```python
# Display classification report
print(classification_report(y_test, pred, target_names=target_names))
```

"False Positive" are almost 100% predicted, but part of that could be caused by imbalanced dataset. "Candidate" has a slightly better precission but worse recall than "Confirmed".

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
lrc_ht = LogisticRegression(random_state=10, penalty='l2', max_iter=1000)
```

```python
# Create the GridSearchCV model
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {
              'solver': ['newton-cg', 'lbfgs', 'sag', 'liblinear'],
              'C': [150, 100, 10, 1.0]
            }
grid = GridSearchCV(lrc_ht, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Define the best estimator, it will be needed for feature importance
# but we can use it instead of grid anywhere else as well
lrg = grid.best_estimator_
```

```python
lrc_rfe = LogisticRegression(random_state=10, penalty='l2', max_iter=1000, C=100, solver='sag')
lrc_rfe.fit(X_train_scaled, y_train)
```

### Recursive Feature Elimination (RFE)

First we will get the ranking of all features by selecting only one feature in RFE

```python
from sklearn.feature_selection import RFE
lrc_rfe = LogisticRegression(random_state=10, penalty='l2', max_iter=1000)
```

```python
# Select only one feature
rfe = RFE(lrc_rfe, n_features_to_select=1)
rfe = rfe.fit(X_train_scaled, y_train)

# List all features with their ranks
sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), feature_names))
```

```python
# Our initial scores before Gridsearch, for comparison
print(f"Training Data Score: {LRC_Train_Score}")
print(f"Testing Data Score: {LRC_Test_Score}")
```

We are going to add features starting from the best ranked feature. In each iteration we will add one more feature and scale, train and score data.

```python
from sklearn.feature_selection import RFE
lrc_rfe = LogisticRegression(random_state=10, penalty='l2', max_iter=1000)

# Start looping
for n in range(1, 21):
    rfe = RFE(lrc_rfe, n_features_to_select=n)
    rfe = rfe.fit(X_train_scaled, y_train)
    
    cols = X_train.columns[rfe.support_]
    
    X_if_train = X_train[cols] 
    X_if_test = X_test[cols]
    y_if_train = y_train
    y_if_test = y_test
    X_if_scaler = MinMaxScaler().fit(X_if_train)
    X_if_train_scaled = X_if_scaler.transform(X_if_train)
    X_if_test_scaled = X_if_scaler.transform(X_if_test)

    lrc_if = LogisticRegression(random_state=10, penalty='l2', max_iter=1000)
    lrc_if.fit(X_if_train_scaled, y_if_train)

    print("=" * 30)
    print(f"Number of selected Features: {n}")
    print("Features sorted by their rank:")
    print(sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), feature_names)))
    ## print(f"Training Data Score: {lrc_if.score(X_if_train_scaled, y_if_train)}")
    print(f"Testing Data Score: {lrc_if.score(X_if_test_scaled, y_if_test)}")
```

First 5 features have the highest importance. Adding each of them is changing score very much. After that the score is changing slightly. Features 19 and 20 are not changing score at all.
